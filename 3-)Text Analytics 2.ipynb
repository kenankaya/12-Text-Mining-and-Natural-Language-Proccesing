# Mathematical Operations and Simple Feature Engineering

```python
text = """
A Scandal in Bohemia! 01
The Red-headed League,2
A Case, of Identity 33
The Boscombe Valley Mystery4
The Five Orange Pips1
The Man with? the Twisted Lip
The Adventure of the Blue Carbuncle
The Adventure of the Speckled Band
The Adventure of the Engineer's Thumb
The Adventure of the Noble Bachelor
The Adventure of the Beryl Coronet
The Adventure of the Copper Beeches"""

text
```

```python
import pandas as pd
v_text = text.split("\n")
v=pd.Series(v_text)
text_vector=v[1:len(v)]
text_vector
```

```python
df=pd.DataFrame(text_vector, columns=["Stories"])
df
```

## 1-)Finding Number of Letters / Characters within the text

```python
c_df=df.copy()
```

```python
c_df
```

```python
c_df["Stories"]
```

```python
c_df["Stories"].str.len()
```

```python
c_df["Number of Chracters"]=c_df["Stories"].str.len()
```

```python
c_df
```

```python
c_df["Number of words "]=c_df["Stories"].apply( lambda x: len(str(x).split(" ")))
```

```python
c_df
```

## 3-)Finding Number of Specials Chracters within the text

```python
c_df["Number of Special Characters-'The' "]=c_df["Stories"].apply(lambda x: len([i for i in x.split() if x.startswith("The")]))
```

```python
c_df
```

## 4-)Finding Number of Digits within the text

```python
c_df["Number of Digit"]=c_df["Stories"].apply(lambda x: len([i for i in x.split() if x.isdigit()]))
```
